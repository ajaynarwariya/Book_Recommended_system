# Collabrating filtering based recommender system

```python
rating_with_books.columns
```

```python
x = rating_with_books.groupby('User-ID').count()['Book-Title']>200
user_rating = x[x].index
```

```python
user_rating
```

```python
filtering_rating = rating_with_books[rating_with_books['User-ID'].isin(user_rating)]
```

```python
filtering_rating.shape
```

```python
rating_with_books.shape
```

```python
y = filtering_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books = y[y].index
famous_books
```

```python
  final_ratings = filtering_rating[filtering_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt= final_ratings.pivot_table(index= 'Book-Title' , columns= 'User-ID' , values = 'Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace= True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
def recommended(book_name):
    #fetch index
    index = np.where(pt.index== book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1], reverse = True)[1:5]
    data= []
    for i in similar_item:
        item = []
        temp_df = books[books['Book-Title']== pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data
    
```

```python
recommended('1st to Die: A Novel')
```
